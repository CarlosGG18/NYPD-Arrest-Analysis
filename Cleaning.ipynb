### Checking for Missing/Duplicate Values

```python
df.isna().sum()
```

```python
missing_percentage = (df['PD_CD'].isnull().sum() / len(df['PD_CD'])) * 100
print("Percentage of missing values in 'PD_CD' column:", missing_percentage)
```

```python
missing_percentage = (df['KY_CD'].isnull().sum() / len(df['KY_CD'])) * 100
print("Percentage of missing values in 'KY_CD' column:", missing_percentage)
```

```python
missing_percentage = (df['LAW_CAT_CD'].isnull().sum() / len(df['LAW_CAT_CD'])) * 100
print("Percentage of missing values in 'LAW_CAT' column:", missing_percentage)
```

```python
misisng_KY_CD_rows = df[df['KY_CD'].isna()]
misisng_KY_CD_rows.head(12)
```

```python
total_rows = len(df)
total_missing_count = df.isnull().sum()

for column, missing_count in total_missing_count.items():
    if missing_count > 0:
        missing_percentage = (missing_count / total_rows) * 100
        print(f"Percentage of missing values in '{column}' column: {missing_percentage:.2f}%")
        
total_missing_percentage = (total_missing_count.sum().sum() / (total_rows * len(df.columns))) * 100
print(f"Total percentage of missing values: {total_missing_percentage:.2f}%")
```
